```python
obj = """
            {"name": "Wes",
            "places_lived": ["United States", "Spain", "Germany"],
            "pet": null,
            "siblings": [{"name": "Scott", "age": 30, "pets": ["Zeus", "Zuko"]},
            {"name": "Katie", "age": 38,
            "pets": ["Sixes", "Stache", "Cisco"]}]
}
"""
```

```python
type(obj)
```

```python
import pandas as pd
import json
result = json.loads(obj)
```

```python
type(result)
```

```python
asjson = json.dumps(result)
```

```python
asjson
```
